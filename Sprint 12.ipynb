### Histogram Kelompok Churn dan Non Churn


```python
# fungsi untuk visualisasi histogram pelanggan churn
def feat_hist(x, group_col, key_cols):
    x_new = x.copy()
    x_new[key_cols] = df[key_cols]
    for c in key_cols:
        for cl in x[group_col].unique():
            sns.histplot(x_new[x_new[group_col]==cl][c],
                         label='{} = {}'.format(group_col, str(cl)), bins = 10, kde=True)
            plt.title(c)
            plt.legend()
            plt.show()
```

```python
feat_hist(df, 'churn', df.columns)
```

Histogram di atas menunjukkan perbedaan distribusi tiap fitur untuk kelompok yang churn dan non-churn. Dalam beberapa fitur, distribusi datanya tidak banyak berbeda. Fitur-fitur di mana distribusi data kedua kelompok ini berbeda adalah: 

**Periode kontrak**    
Variasi periode kontrak untuk kelompok churn lebih sedikit. Kebanyakan memiliki kontrak 1 hingga 3 bulan. 

**Usia**
Variasi usia pada kelompok churn juga lebih sedikit. Rata-rata usia kelompok ini adalah **26 tahun**, sedangkan kelompok non-churn rata-rata berusia **29 tahun**. 

**Lifetime**
Variasi lama keanggotaan pada kelompok churn juga lebih kecil. Kebanyakan pelanggan churn telah menjadi anggota selama **3 bulan** atau kurang, terhitung sejak mendaftar. 

**Rata-rata frekuensi kelas secara total**

## Model Prediksi Churn Pelanggan

Bagian ini berfokus pada klasifikasi biner dengan target variabel status 'churn' pada pelanggan. Data akan dibagi menjadi 'training data' dan 'validation data' secara acak, dengan proporsi 80:20. Lalu dataset 'training data' akan dilatih menggunakan dua model: **regresi logistik** dan **random forest**. 
Bagian ini akan diakhiri dengan metrik evaluasi terhadap kedua model, untuk menentukan model terbaik dalam memprediksi tingkat churn pelanggan. 

### Pembagian Data

```python
# pembagian data training dan validation
X = df.drop(columns=['churn'])
y = df['churn']

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=2022)
```

```python
X_train.head()
```

```python
X_val.head()
```

### Melatih Model

Diberlakukan dua metode untuk melatih model, yaitu regresi logistik dan random forest. Keduanya diharapkan melatih model untuk mendeteksi nilai true positive sebanyak mungkin dan false positive sesedikit mungkin. Kinerja kedua metode akan dievaluasi dengan metrik Accuracy, Precision dan Recall. 

#### Regresi Logistik

```python
# Melatih model dengan regresi logistik 
lr_model = LogisticRegression(random_state=0)

lr_model.fit(X_train, y_train)

y_proba_lr = lr_model.predict_proba(X_val)
y_pred_lr = lr_model.predict(X_val)
y_pred_train_lr = lr_model.predict(X_train)
```

```python
# Visualisasi dengan kurva roc
# Confusion Matrix
cm = confusion_matrix(y_val, y_pred_lr)
print("Confusion Matrix:")
print(cm)

# ROC Curve
fpr, tpr, thresholds = roc_curve(y_val, y_proba_lr[:, 1])
plt.plot(fpr, tpr, label='ROC Curve (AUC = {:.2f})'.format(roc_auc_score(y_val, y_proba_lr[:, 1])))
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

Metode regresi logistik telah melatih model untuk mengidentifikasi banyak poin dengan nilai True Positive tinggi dan nilai False Positive rendah.

```python
# Visualisasi kurva Precision-Recall
precision, recall, thresholds = precision_recall_curve(y_val, y_proba_lr[:, 1])
plt.plot(recall, precision, label='Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.show()
```

Terdapat nilai precision dan recall yang sama-sama tinggi, yang mengindikasikan bahwa model ini mampu mengidentifikasi sebagian besar kasus positif dalam dataset, dengan hanya membuat sedikit kesalahan prediksi (prediksi false positive).
